## **Data Preprocessing and Exploration**

```python
df.isna().sum()
```

* **There is only 3 null values, In the column of Financial Stress**

* **So the percentage of nullvalue must be very low**

* **Fill them by using fillna**

```python
df.isna().sum()*100/len(df)
```

```python
df.fillna(method = 'ffill',inplace = True)
```

```python
df.duplicated().sum()
```

* **There is no duplicat values**

```python
df.drop('id',axis=1,inplace=True)
```

* **In this prediction ID has no role so we can delete it**

---

```python
for i in df.columns:
  sns.boxplot(df[i])
  plt.show()
  print('\n')
```

* **There is no Outliers**

```python
for i in df.columns:
  print(i,df[i].unique())
  print('-'*100)
  print(i,df[i].value_counts())
  print('*'*100)
```

```python
df.rename(columns={'Have you ever had suicidal thoughts ?': 'Suicid Attempt'},inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le.fit(df['City'])
df['City'] = le.transform(df['City'])

le1 = LabelEncoder()
le1.fit(df['Profession'])
df['Profession'] = le1.transform(df['Profession'])

le2 = LabelEncoder()
le2.fit(df['Sleep Duration'])
df['Sleep Duration'] = le2.transform(df['Sleep Duration'])

le3 = LabelEncoder()
le3.fit(df['Degree'])
df['Degree'] = le3.transform(df['Degree'])

le4 = LabelEncoder()
le4.fit(df['Suicid Attempt'])
df['Suicid Attempt'] = le4.transform(df['Suicid Attempt'])

le5 = LabelEncoder()
le5.fit(df['Family History of Mental Illness'])
df['Family History of Mental Illness'] = le5.transform(df['Family History of Mental Illness'])
```

* **Lebelencoded 7 columns for transform object data to numeric**

* **There are too many categories so we can use labelencoding**

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse_output=False)
ohe.fit(df[['Dietary Habits']])
col = ohe.get_feature_names_out(['Dietary Habits'])
df1 = pd.DataFrame(ohe.transform(df[['Dietary Habits']]),columns=col)

ohe1 = OneHotEncoder(sparse_output=False)
ohe1.fit(df[['Gender']])
col = ohe1.get_feature_names_out(['Gender'])
df2 = pd.DataFrame(ohe1.transform(df[['Gender']]),columns=col)
```

* **OneHotEncoded 'Dietary Habits'and 'Gender'**

```python
df = pd.concat([df,df1,df2],axis=1)
df.drop(['Dietary Habits','Gender'],axis=1,inplace=True)
```
